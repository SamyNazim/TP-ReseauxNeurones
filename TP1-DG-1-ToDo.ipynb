## 1. IMPORT DES BIBLIOTHÈQUES

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. DATASET

```python
# y = 2x + 1 + bruit
np.random.seed(42)

X = np.linspace(0, 10, 50)
bruit = np.random.randn(50) * 1.5
y = 2 * X + 1 + bruit
```

## 3. VISUALISATION DES DONNÉES

```python
plt.scatter(X, y)
plt.title("Dataset initial")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

## 4. INITIALISATION DES PARAMÈTRES

```python
theta_0 = 0.0   # biais
theta_1 = 0.0   # pente

learning_rate = 0.01
epochs = 10
```

## 5. FONCTIONS DE LA DESCENTE DU GRADIENT

```python
def prediction(x, theta_1, theta_0):
    return theta_0 + theta_1 * x


def cost_function(X, y, theta_1, theta_0):
    m = len(X)
    predictions = theta_0 + theta_1 * X
    error = predictions - y
    J = (1/(2*m)) * np.sum(error**2)
    return J


def gradients(X, y, theta_1, theta_0):
    m = len(X)
    
    predictions = theta_0 + theta_1 * X
    error = predictions - y
    
    d_theta_0 = (1/m) * np.sum(error)
    d_theta_1 = (1/m) * np.sum(error * X)
    
    return d_theta_1, d_theta_0
```

```python
# Desente du gradient
cost_history = []
cost = cost_function(X, y, theta_1, theta_0)

for epoch in range(epochs):
    d_theta_1, d_theta_0 = gradients(X, y, theta_1, theta_0)
    
    theta_1 = theta_1 - learning_rate * d_theta_1
    theta_0 = theta_0 - learning_rate * d_theta_0
    
    cost = cost_function(X, y, theta_1, theta_0)
    cost_history.append(cost)
```

```python
# Affichage des paramètres finaux
print("\nParamètres appris :")
print("theta_1 =", theta_1)
print("theta_0 =", theta_0)
```

## 6. VISUALISATION 

```python
plt.plot(cost_history)
plt.title("Évolution du coût")
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.show()
```
